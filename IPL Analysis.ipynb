# Extracting the records where a team won  batting first.

```python
batting_first.head()
```

# Making A bar plot for top 3 teams with most wins after batting first.

```python
plt.figure(figsize=(6, 8))

# Assuming "batting_first" is your DataFrame
winner_counts = batting_first['winner'].value_counts()
top_three_winners = winner_counts.head(3)

plt.bar(top_three_winners.index, top_three_winners.values, color=["Blue", "Green", "Red"])
plt.show()
```
